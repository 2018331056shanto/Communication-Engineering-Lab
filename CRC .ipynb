## How it works

**input data:** 11010011101100

This is first padded with zeros corresponding to the bit length n of the CRC. This is done so that the resulting code word is in systematic form. Here is the first calculation for computing a 3-bit CRC:

#### The algorithm acts on the bits directly above the divisor in each step. The result for that iteration is the bitwise XOR of the polynomial divisor with the bits above it. The bits not above the divisor are simply copied directly below for that step. The divisor is then shifted right to align with the highest remaining 1 bit in the input, and the process is repeated until the divisor reaches the right-hand end of the input row.
```python
#CRC Clyclic Redundency Check
divisor=input("Enter Divisor :")
dividend=input("Enter Dividend :")
divisor1=[]
dividend1=[]
store=[]
def calculate_CRC(divisor,dividend):
  size=len(dividend)
  sizediv=len(divisor)
  cnt=0

  while cnt+sizediv<=size:
    
    start=0
    count=0

    for x in range(0,size,1):
      count+=1

      if dividend[x]==1:
        start=x
        break

        if count>size:
          break ;
    
    idx=0
    temp=start
    cnt=start

    if cnt+sizediv>size:
      break;
    for x in range(start,start+sizediv,1):
      
      if dividend[x]==divisor[idx]:
        dividend[x]=0

      else:
        dividend[x]=1
  
      idx+=1
      
  return dividend
     


idx=0

for x in divisor:
  divisor1.insert(idx,int(x))
  idx+=1

idx=0

for x in dividend:

  dividend1.insert(idx,int(x))
  store.insert(idx,int(x))
  idx+=1

for x in range(0,len(divisor1)-1,1):
  dividend1.insert(len(dividend1)+x,0)
  store.insert(len(store)+x,0)


ans=calculate_CRC(divisor1,dividend1)
ans=ans[len(dividend):]
idx=0

for x in range(len(dividend),len(store),1):
  store[x]=ans[idx]
  idx+=1
print("Code word received at the receiver side :",store)
```
